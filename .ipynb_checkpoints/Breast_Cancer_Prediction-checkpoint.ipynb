# Breast Cancer Prediction

## Attributes Info.
1. id - ID number

2. diagnosis - The diagnosis of breast tissues (M = malignant(Cancerous), B = benign(Non Cancerous))

3. radius_mean - Mean of distances from center to points on the perimeter

4. texture_mean - Standard deviation of gray-scale values

5. perimeter_mean - Mean size of the core tumor

6. area_mean

7. smoothness_mean - Mean of local variation in radius lengths

8. compactness_mean - Mean of perimeter^2 / area - 1.0

9. concavity_mean - Mean of severity of concave portions of the contour

8. concave points_mean - Mean for number of concave portions of the contour

10. symmetry_mean

11. fractal_dimension_mean - Mean for "coastline approximation" - 1

12. radius_se - Standard error for the mean of distances from center to points on the perimeter

13. texture_se - Standard error for standard deviation of gray-scale values

14. perimeter_se

15. area_se

16. smoothness_se - Standard error for local variation in radius lengths

17. compactness_se - Standard error for perimeter^2 / area - 1.0

18. concavity_se - Standard error for severity of concave portions of the contour

19. concave points_se - Standard error for number of concave portions of the contour

20. symmetry_se

21. fractal_dimension_se - Standard error for "coastline approximation" - 1

22. radius_worst - Worst or largest mean value for mean of distances from center to points on the perimeter

23. texture_worst - Worst or largest mean value for standard deviation of gray-scale values

24. perimeter_worst

25. area_worst

26. smoothness_worst - Worst or largest mean value for local variation in radius lengths

27. compactness_worst - Worst or largest mean value for perimeter^2 / area - 1.0

28. concavity_worst - Worst or largest mean value for severity of concave portions of the contour

29. concave points_worst - Worst or largest mean value for number of concave portions of the contour

30. symmetry_worst

31. fractal_dimension_worst - Worst or largest mean value for "coastline approximation" - 1

```python
from sklearn.model_selection import train_test_split
```

```python
def splitting(XX, YY):
    X_train, X_test, y_train, y_test = train_test_split(XX, YY, test_size=0.30, random_state=0)
    return X_train, y_train, X_test, y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn import metrics
from statistics import mean
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
def classification_model(model, X_Train, y_Train, X_Test, y_Test):
    model.fit(X_Train, y_Train)
    y_pred = model.predict(X_Test)
    model_score = accuracy_score(y_Test, y_pred)*100
    return round(model_score,2)
```

```python
def cross_validation(X_dataset, clf_model):
    cv = KFold(n_splits = 10, shuffle = False, random_state = 0)
    acc_cv = []
    for train_index, test_index in cv.split(X_dataset):
        X_train, X_test = X_dataset[train_index], X_dataset[test_index]
        y_train, y_test = y[train_index], y[test_index]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc_cv.append(accuracy_score(y_test, y_pred))
    return mean(acc_cv)
```

```python
models = []
models.append(('DT', DecisionTreeClassifier(random_state = 0)))
#models.append(('NB', GaussianNB()))
models.append(('LR', LogisticRegression(random_state = 0)))
#models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('SVM', SVC(random_state = 0)))
models.append(('RF', RandomForestClassifier(random_state = 0)))
names = []
pre_acc = []
for name, model in models:
    mean_acc_cv = cross_validation(X_scaled, model)
    mean_acc_cv = round((mean_acc_cv*100), 2)
    pre_acc.append(mean_acc_cv)
    print(name,':', mean_acc_cv)
```

```python
from sklearn.model_selection import GridSearchCV
```

# Hyper-Parameter Tuning for Decsion Tree

```python
with warnings.catch_warnings():
    warnings.filterwarnings(action='ignore',category=DeprecationWarning)
    criterion = ['gini', 'entropy']
    max_depth = list(range(1,10))
    min_samples_leaf = list(range(1,10))
    max_leaf_nodes = list(range(2,10))
    random_state = [0,42]
    hyperparameter = dict(criterion = criterion, max_depth = max_depth, min_samples_leaf = min_samples_leaf, random_state = random_state)
    clf_dt = DecisionTreeClassifier()
    grid_dt = GridSearchCV(clf_dt, hyperparameter, cv = 10, n_jobs= -1)
    X_train, y_train, X_test, y_test = splitting(X_scaled_df, y)
    grid_dt.fit(X_train, y_train)
```

```python
print('Best parameter: ', grid_dt.best_estimator_)
print('Best score:', grid_dt.best_score_)
```

```python
X_train, y_train, X_test, y_test = splitting(X_scaled_df, y)
acc_dt = classification_model(grid_dt.best_estimator_, X_train, y_train, X_test, y_test)
print('Accuracy using Decision Tree:', acc_dt,'%')
```

# Hyper-Parameter Tuning for Logistic Regression

```python
with warnings.catch_warnings():
    warnings.filterwarnings(action='ignore',category=DeprecationWarning)
    C = [0.001,0.01,0.1,1,10,100,1000]
    random_state = [0, 42]
    penalty = ['l1', 'l2']
    hyperparameter = dict(C= C, penalty = penalty, random_state = random_state)
    clf_lr = LogisticRegression()
    grid_lr = GridSearchCV(clf_lr, hyperparameter, cv = 10, n_jobs= -1)
    X_train, y_train, X_test, y_test = splitting(X_scaled_df, y)
    grid_lr.fit(X_train, y_train)
```

# Hyper-Parameter Tuning for Support Vector Machine

```python
with warnings.catch_warnings():
    warnings.filterwarnings(action='ignore',category=DeprecationWarning)
    C = [0.1, 1, 10, 100, 1000]
    kernel = ['linear','rbf', 'sigmoid']
    gamma = [1, 0.1, 0.01, 0.001, 0.0001]
    random_state = [0, 42]
    hyperparameter = dict(C = C, kernel = kernel, gamma = gamma, random_state = random_state)
    clf_svm = SVC()
    grid_svm = GridSearchCV(clf_svm, hyperparameter, cv = 10, n_jobs= -1)
    X_train, y_train, X_test, y_test = splitting(X_scaled_df, y)
    grid_svm.fit(X_train, y_train)
```

# K Means 

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
pca = PCA(n_components = 2) 
X_pca = pca.fit_transform(X_scaled_df)
cluster_no = range(2, 8)
acc_k_means = []
for i in cluster_no:
    k_means = KMeans(n_clusters= i, random_state=0)
    pred_target = k_means.fit_predict(X_pca)
    print(i, ':', round((accuracy_score(y, pred_target)*100), 2))
    acc_k_means.append(accuracy_score(y, pred_target)*100)
plt.figure(figsize = (10,5))
plt.plot(cluster_no, acc_k_means)
plt.xlabel('No of cluster')
plt.ylabel('Accuracy')
```

```python
k_means = KMeans(n_clusters= 2, random_state=0)
pred_target = k_means.fit_predict(X_pca)
original_target = pd.DataFrame(np.array(y))
predicted_target = pd.DataFrame(pred_target)
k_means_df = pd.concat([original_target, predicted_target], axis= 1)
k_means_df.columns = ['Original_target', 'Predicted_target']
k_means_df['Decision'] = np.where(k_means_df['Original_target'] == k_means_df['Predicted_target'], 'True', 'False')
k_means_df.head(n=12)
```

```python
print('Accuracy using K Means:', round((accuracy_score(original_target, predicted_target)*100),2),'%')
```
